## 4. Data Preprocessing

<a id=section401></a>
### 4.1 Data Preprocessing

- Here we will perform **Data Preprocessing** on our dataset, to make the data usable for **EDA**.

````python
pre_movies.head()
````

- **Rating** column is multiplied with 10 to **standardise** it with **Metascore** column

````python
pre_movies['Rating'] = pre_movies['Rating'] * 10
````

````python
pre_movies.describe()
````

- **Revenue** column having **Zero value** is for the movie **A Kind of Murder** which is impossible, Thus should be considered as Null value as well.

````python
pre_movies['Revenue (Millions)'] = pre_movies['Revenue (Millions)'].replace(0.0, np.nan)
````

````python
pre_movies.describe()
````

- Rounding the **Revenue** column to 3 decimal places.

````python
pre_movies['Revenue (Millions)'] = pre_movies['Revenue (Millions)'].round(decimals=2)
````

````python
pre_movies.describe()
````

- Treating the **missing values** of **Revenue & Metascore** columns with **KNNImputer**

````python
#!pip install -U scikit-learn
````

````python
from sklearn.impute import KNNImputer
````

````python
num_missing = (pre_movies[['Revenue (Millions)', 'Metascore']].isna()).sum()
````

````python
num_missing
````

````python
test_movies = pre_movies
````

````python
values = test_movies.iloc[:, 7:].values
````

````python
values
````

````python
# Count the number of NaN values in each column
print('Missing: %d' % np.isnan(values).sum())
````

````python
# Define the imputer
knn_imputer = KNNImputer(n_neighbors=3, weights="uniform")
````

````python
# Transform the dataset
transformed_values = knn_imputer.fit_transform(values)
````

````python
# Count the number of NaN values in each column
print('Missing: %d' % np.isnan(transformed_values).sum())
````

````python
type(transformed_values)
````

````python
knn_movies = pd.concat([test_movies.iloc[:, :-5], pd.DataFrame(transformed_values)], axis=1)
````

````python
knn_movies.columns
````

````python
knn_movies.rename(columns = {0:'Runtime (Minutes)', 1:'Rating', 2:'Votes', 3:'Revenue (Millions)', 4:'Metascore'}, inplace = True) 
````

````python
knn_movies.describe()
````

- Creating **new columns** for **seperate Genres** as **boolean values**

````python
GenreList = ['Sci-Fi', 'Western', 'Comedy', 'Action', 'Family', 'Animation', 'Biography', 'Horror', 
             'History', 'Crime', 'Sport', 'Adventure', 'Mystery', 'Thriller', 'War', 'Romance', 'Fantasy', 'Drama', 'Music']
````

````python
data = {}

for Genre in GenreList:
    data[Genre] = []
````

````python
for rank in range(1,1001):
    for Genre in GenreList:
        data[Genre].append(knn_movies[knn_movies['Rank'] == rank]['Genre'].str.contains(Genre).iloc[0])
````

````python
genre_df = pd.DataFrame.from_dict(data)
genre_df.head()
````

````python
genre_df = genre_df.astype('int64')
````

````python
movie_df = pd.concat([knn_movies, genre_df], axis=1, sort=False)
````

## Deriving Another Genre Dataset

````python
Gendata = {}

Gendata['Genre'] = []
Gendata['Revenue'] = []
Gendata['Year'] = []
Gendata['User Rating'] = []
Gendata['Critic Rating'] = []
Gendata['Runtime'] = []
Gendata['Votes'] = []
````

````python
GenreList = ['Sci-Fi', 'Action', 'Adventure', 'Mystery', 'Horror', 'Thriller','Comedy', 'Family', 'Animation', 'Fantasy', 'Drama', 'Music',
             'Biography', 'Romance', 'History','Crime', 'Western', 'War', 'Sport']
````

````python
for rank in range(1,1001):
    
    for Genre in GenreList:
        Gendata['Genre'].append(Genre)
        Gendata['Revenue'].append(movie_df[(movie_df['Rank'] == rank) & (movie_df['Genre'].str.contains(Genre))]['Revenue (Millions)'].sum())
        Gendata['User Rating'].append(movie_df[(movie_df['Rank'] == rank) & (movie_df['Genre'].str.contains(Genre))]['Rating'].sum())
        Gendata['Critic Rating'].append(movie_df[(movie_df['Rank'] == rank) & (movie_df['Genre'].str.contains(Genre))]['Metascore'].sum())
        Gendata['Runtime'].append(movie_df[(movie_df['Rank'] == rank) & (movie_df['Genre'].str.contains(Genre))]['Runtime (Minutes)'].sum())
        Gendata['Votes'].append(movie_df[(movie_df['Rank'] == rank) & (movie_df['Genre'].str.contains(Genre))]['Votes'].sum())
        Gendata['Year'].append(movie_df[(movie_df['Rank'] == rank) & (movie_df['Genre'].str.contains(Genre))]['Year'].sum())
````

````python
genredist_df = pd.DataFrame.from_dict(Gendata)
````

````python
genredist_df = genredist_df[genredist_df['Revenue'] != 0]
````

````python
genredist_df[['Runtime', 'User Rating', 'Critic Rating', 'Votes']] = genredist_df[['Runtime', 'User Rating', 'Critic Rating', 'Votes']].astype('int64')
genredist_df['Revenue'] = genredist_df['Revenue'].round(decimals=2)
````

````python
genredist_df.to_csv('genredist.csv', index=False)
````

## 5. Exploratory Data Analysis

**Exploratory Data Analysis(EDA)** is an **approach to analyzing data sets** to **summarize** their **main characteristics**, often with **visual methods**. 


- It includes **cleaning, munging, combining, reshaping, slicing, dicing, and transforming data** for **analysis** purpose.


- The **primary goal of EDA** is to **maximize the analyst's insight into a data set** and into the **underlying structure of a data set**, while **providing all of the specific items** that an **analyst would want to extract from a data set**, such as:
 <br><br> 
  - A list of **outliers**.<br><br>
  
  - A **good-fitting model**.  <br><br>
  
  - **Estimates for parameters**.<br><br>
  
  - **Uncertainties** for those **estimates**.<br><br>
  
  - A **ranked list** of **important factors**.<br><br>
  
  - **Conclusions** as to whether **individual factors are statistically significant**.<br><br>
  
  - A **sense of robustness of conclusions**.<br><br>
  
  - **Optimal settings**.<br><br>

### 5.1.1 How is the movie industries growing over time?

````python
movieVSyear = sns.catplot(x="Year", kind="count", palette="Blues", height=7, aspect=2, data=movie_df)

movieVSyear.set_xticklabels(rotation=45)
movieVSyear.set_ylabels("Number of Movies")

sns.set_style("whitegrid", 
              {'xtick.bottom': True})

Years = movie_df['Year'].unique().tolist()
Years.sort()

for x in range(0, len(Years)):
    movieVSyear.ax.text(x, 
                        movie_df['Year'].value_counts()[Years[x]]-15, 
                        str(movie_df['Year'].value_counts()[Years[x]]) + " Nos", 
                        ha='center', 
                        color='white')

    
movieVSyear.ax.annotate('Highest Movies Made', 
                        xy=(10, 297), xytext=(7, 250),
                        arrowprops=dict(facecolor='red', shrink=0.05))

plt.suptitle('Bar chart showing the growth of Movie Industry [2006 - 2016]', y=1.05, fontsize=18, ha='center')
plt.title('(Higher is better)', fontsize=10, ha='center')
````

- There has been a **linear increase** in number of **movies produced** from **2006 to 2016**
- Where **2016** produces the **highest of 297 movies** in a year.

### 5.5.1 Top 3 Directors of 9 Filtered Genres?

````python
from math import pi
````

````python
def GenreRadar(Genre, cmap):

    radardict = {}
    radardict['group'] = movie_df[movie_df['Genre'].str.contains(Genre)][['Director', 'Revenue (Millions)']].groupby(['Director'])['Revenue (Millions)'].sum().sort_values(ascending=False).head(3).index.tolist()
    radardict['Revenue'] = movie_df[movie_df['Genre'].str.contains(Genre)][['Director', 'Revenue (Millions)']].groupby(['Director'])['Revenue (Millions)'].sum().sort_values(ascending=False).head(3).tolist()
    radardict['Runtime'] = []
    radardict['Rating'] = []
    radardict['Metascore'] = []
    radardict['Votes'] = []
    
    for director in movie_df[movie_df['Genre'].str.contains(Genre)][['Director', 'Revenue (Millions)']].groupby(['Director'])['Revenue (Millions)'].sum().sort_values(ascending=False).head(3).index.tolist():
        radardict['Runtime'].append(movie_df[(movie_df['Genre'].str.contains(Genre) & (movie_df['Director'].str.contains(director)))][['Director', 'Runtime (Minutes)']].groupby(['Director'])['Runtime (Minutes)'].mean().tolist()[0])
        radardict['Rating'].append(movie_df[(movie_df['Genre'].str.contains(Genre) & (movie_df['Director'].str.contains(director)))][['Director', 'Rating']].groupby(['Director'])['Rating'].mean().tolist()[0])
        radardict['Metascore'].append(movie_df[(movie_df['Genre'].str.contains(Genre) & (movie_df['Director'].str.contains(director)))][['Director', 'Metascore']].groupby(['Director'])['Metascore'].mean().tolist()[0])
        radardict['Votes'].append(movie_df[(movie_df['Genre'].str.contains(Genre) & (movie_df['Director'].str.contains(director)))][['Director', 'Votes']].groupby(['Director'])['Votes'].mean().tolist()[0] / 1000)
 
    # Set data
    df = pd.DataFrame(radardict)
    
    # ------- PART 2: Apply to all individuals
    # initialize the figure
    my_dpi=96
    plt.figure(figsize=(1000/my_dpi, 1000/my_dpi), dpi=my_dpi)

    # Create a color palette:
    my_palette = plt.cm.get_cmap(cmap, len(df.index))
    
    # Loop to plot
    for row in range(0, len(df.index)):
        make_spider( row=row, title='Director - '+df['group'][row], color=my_palette(row))
    print(df)
````

````python
# ------- PART 1: Define a function that do a plot for one line of the dataset!

def make_spider( row, title, color):

    # number of variable
    categories=list(df)[1:]
    N = len(categories)

    # What will be the angle of each axis in the plot? (we divide the plot / number of variable)
    angles = [n / float(N) * 2 * pi for n in range(N)]
    angles += angles[:1]

    # Initialise the spider plot
    ax = plt.subplot(1,3,row+1, polar=True, )

    # If you want the first axis to be on top:
    ax.set_theta_offset(pi / 2)
    ax.set_theta_direction(-1)

    # Draw one axe per variable + add labels labels yet
    plt.xticks(angles[:-1], categories, color='grey', size=8)



    # Ind1
    values=df.loc[row].drop('group').values.flatten().tolist()
    values += values[:1]
    ax.plot(angles, values, color=color, linewidth=2, linestyle='solid')
    ax.fill(angles, values, color=color, alpha=0.4)

    # Add a title
    plt.title(title, size=11, color=color, y=1.1)
    
````
